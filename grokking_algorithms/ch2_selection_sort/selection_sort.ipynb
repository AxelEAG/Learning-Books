For a list of _n_ elements, the algorithm must scan the unsorted portion for each element—roughly _n_ times for _n_ elements—giving a time complexity of _O(n × n) = O(n²)_.

One might note: it doesn't actually make _n_ comparisons each time. As elements move to the sorted section, the unsorted portion shrinks: _n_, then _n–1_, _n–2_, … down to 1. On average, it performs about _n/2_ comparisons for each of the _n_ iterations, or _n²/2_ in total. However, big-O notation ignores constants, so the overall complexity remains _O(n²)_.
